## String Functions
- Case conversion function - lower(), upper(), title(), sewapcase(), capitalize()

```python
s = "I am PytHon Batch studEnT"
print(s.lower())
print(s.upper())
print(s.title())
print(s.swapcase())
print(s.capitalize())
```

## Sub string index finding
- find()


```python
# find()
s = "I am python batch student"
print(s.find("t"))
print(s.find("t", 10)) # it stats checking from 10th position
print(s.find('z'))
print(s.index('t'))
```

- strip()

```python
# strip() - it removes the white spaces
s = "         codegnan       "
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s = "@@@@@code@@gnan@@@@"
print(s.strip('@'))
print(s.lstrip('@'))
print(s.rstrip())
```

- split()

```python
s = "I am python batch student"
print(s.split())
```

## Username validation

```python
username = input("Enter the username: ")
if len(username) >=8 and username[0].isalpha():
    for char in username:
        if char.isupper():
            has_upper=True
        elif char.islower():
            has_lower=True
        elif char.isdigit():
            has_digit=True
        elif char in ["_","@"]:
            has_special=True
    if has_upper and has_lower and has_digit and has_special:
        print("Username is valid!")
    else:
        print("Username is invalid!")               
else:
    print("Username is invalid!")
    
```

## Password validation

```python
password = input("Enter the password: ")
if len(password) >=8 and password[0].isupper():
    for char in password:
        if char.isupper():
            has_upper=True
        elif char.islower():
            has_lower=True
        elif char.isdigit():
            has_digit=True
        elif char in ["_","@"]:
            has_special=True
    if has_upper and has_lower and has_digit and has_special:
        print("Password is valid!")
    else:
        print("Username is invalid!")
else:
    print("Username is invalid!")
```
