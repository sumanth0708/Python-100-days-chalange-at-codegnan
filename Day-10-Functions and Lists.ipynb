1.Positional Arguments

```python
# Subtraction of two numbers using functions
def sub(n1, n2):
    res = n1 - n2
    return res
a = int(input())
b = int(input())
print(sub(a, b))
```

2.Keyword Arguments

```python
# Subtraction of two numbers using functions
def sub(n1, n2):
    res = n1 - n2
    return res
a, b = 10, 5
print(sub(n1 = a,n2 = b))
print(sub(n2 = b,n1 = a))
```

```python
# Sum of 2 numbers using functions
def sum(n1, n2):
    res = n1 + n2
    return res
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print(sum(a, b))
```

```python
# Sum of 3 numbers using functions
def sum(n1, n2, n3):
    res = n1 + n2 + n3
    return res
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))
print(sum(a, b, c))
```

## List Properties

```python
#List elements access by using index position\
lst = [10,2,4,4.534, 'hi', 'hello']
print(lst[2])
print(lst[-3])
print(lst[2:])
```

```python
# nested list element access
lst = [1,2,3,4,[10,30,50,['a','b','c']],29,1,3]
print(lst[4])
print(lst[4][1])
print(lst[4][3][2])
```

```python
#List is mutable
lst = [1,2,3,40, 'hi', 'hello']
lst[0] =100
print(lst)
```

## Map function

```python
# Reading list of integer elements from user
l = list(map(int, input().split()))
print(type(l),l)
```

```python
# read List of number and if number is even add 0.5 and if number is odd add 1
# Lst = [1,2,3,4,5]
# res = [2, 2.5, 4, 4.5, 6]
lst = list(map(int, input().split()))
for i in range(0, len(lst)):
    if lst[i] % 2:
        lst[i] = lst[i] + 1
    else:
        lst[i] += 0.5
print(lst)
```
