```python
# nested Loops
for i in range(1, 3):
    for j in range(1, 11):
        print(f"{i} * {j} = {i*j}")
    print("____________________")
```

```python
# write 3 table program in functions
def print_table(n):
    for i in range(1, 11):
        print(f'{n} * {i} = {n*i}')

# 3 table
print_table(3)
# print_table(4)
```

```python
# write a table using functions
def print_table(n):
    for i in range(1, 11):
        print(f'{n} * {i} = {n*i}')

n = int(input("Enter any number: "))
print_table(n)
```

```python
# print n stars in horigentally
n = int(input())
for i in range(1, n+1):
    print("*", end= ' ')
```

```python
# rectangular pattern 
n = int(input())
for i in range(1, n+1):
    for j in range(1, n+1):
        print("*", end= ' ')
    print()
```
