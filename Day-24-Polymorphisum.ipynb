```python
# method overriding
# person
class person():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def show(self):
        return f"Myself {self.name} and I am a {self.age} years old"

#student
class student(person):
    def __init__(self, name, age, std):
        # super().__init__(name, age)
        person.__init__(self, name, age)
        self.std = std

    def show(self):
        return f"Myself {self.name} and I am studying {self.std} class"

p = person('Babu', 50)
s = student('srinu', 15, 10)
print(p.show())
print(s.show())
        
```

```python
# Method overriding
# person
class person():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        return f"Myself {self.name} and I am a {self.age} years old"

#student
class student(person):
    def __init__(self, name, age, std):
        # super().__init__(name, age)
        person.__init__(self, name, age)
        self.std = std

    def show(self):
        print(super().show())
        return f"Myself {self.name} and I am studying {self.std} class"

p = person('Babu', 50)
s = student('srinu', 15, 10)
# print(p.show())
print(s.show())
        
```

```python
# Method overloading
class Math:
    def add(self, a, b):
        return a+b
    def add(self, a, b, c):
        return a+b+c
    def add(self, a, b, c, d):
        return a+b+c+d
m = Math()
print(m.add(1, 2, 3, 4))
print(m.add(1, 2, 3))
print(m.add(1, 2))
```

```python
# Method overloading by using default arguments
class Math:
    def add(self, a, b, c=0, d=0):
        return a+b+c+d
m = Math()
print(m.add(1, 2, 3, 4))
print(m.add(1, 2, 3))
print(m.add(1, 2))
```

```python
# Method overloading by using default arguments
class Math:
    def add(self, a, b, c=0, d=0, *args):
        print(type(args))
        return a+b+c+d
m = Math()
print(m.add(1, 2, 3, 4, 5, 6, 7))
print(m.add(1, 2, 3))
print(m.add(1, 2))
```
