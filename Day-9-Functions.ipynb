## Username validation

```python
s = input("Enter the Username: ")
validate_string = "qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM1234567890@_"
if len(s) >= 8:
    # check atleast one upper case and one lower case letter
     # first character is upper or lower
    if s[0].isupper() or s[0].islower():
        for i in range(1, len(s)):
            if s[1] not in validate_string:
                print(f"Invalid username because character is {s[i]}")
                break
        else:
            print("Valid string")
            # special characters ('_','@')
            # it contains digits
    else:
        print("Username invalid because first character is not upper or not lower")
else:
    print("Username invalid because the length is more than 9") 
```

## Loop control statements
- break

```python
for i in range(1,6):
    if i==5:
        break
    print(i)
else:                        #else block is executed when break is not executed 
    print("program done")
```

- continue

```python
for i in range(1,6):
    if i==3:
        continue
    print(i)
else:
    print("program done")
```
