```python
class Book:
    def __init__(self, book_id, title, author, price):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.price = price

    def display(self):
        print(f"{self.book_id}\t{self.title}\t{self.author}\t{self.price}")


class BookManager:
    def __init__(self):
        self.books = []

    def add_book(self):
        book_id = input("Enter Book ID: ")
        title = input("Enter Title: ")
        author = input("Enter Author: ")
        price = float(input("Enter Price: "))
        
        book = Book(book_id, title, author, price)
        self.books.append(book)
        print("Book added successfully!\n")

    def display_books(self):
        if not self.books:
            print("No books available.\n")
            return
        
        print("\nID\tTitle\tAuthor\tPrice")
        print("-" * 40)
        for book in self.books:
            book.display()
        print()

    def search_book(self):
        search_id = input("Enter Book ID to search: ")
        for book in self.books:
            if book.book_id == search_id:
                print("\nBook Found:")
                print("ID\tTitle\tAuthor\tPrice")
                book.display()
                return
        print("Book not found!\n")

    def delete_book(self):
        del_id = input("Enter Book ID to delete: ")
        for book in self.books:
            if book.book_id == del_id:
                self.books.remove(book)
                print("Book deleted successfully!\n")
                return
        print("Book not found!\n")


# Main Program
manager = BookManager()

while True:
    print("====== Book Management System ======")
    print("1. Add Book")
    print("2. Display All Books")
    print("3. Search Book")
    print("4. Delete Book")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        manager.add_book()
    elif choice == '2':
        manager.display_books()
    elif choice == '3':
        manager.search_book()
    elif choice == '4':
        manager.delete_book()
    elif choice == '5':
        print("Thank you!.")
        break
    else:
        print("Invalid choice! Try again.\n")
```
