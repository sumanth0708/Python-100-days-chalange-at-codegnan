```python
details = {'name':'srinu', 'batch':47, 'course':'Python'}
print(type(details))
print(details)
```

```python
# access distionary values
details = {'name':'srinu', 'batch':47, 'course':'Python'}
print(details['name'])
#print(details['Name']) #if key is not in dictionary it raises Error
```

```python
# Updating the dictionary values
details = {'name':'srinu', 'batch':47, 'course':'Python'}
details['name'] = 'babu'
print(details)
```

```python
# Updating the dictionary values
details = {'name':'srinu', 'batch':47, 'course':'Python'}
details['name'] = 'babu'
print(details)
# Assigning new item into dictionary
# dict_obj[key] = values
details['age'] = 50
print(details)
```

```python
## Dictionary functions
# accessing values
# get(key, default_values): it returns value if key is present in dictionary otherwise default value
details = {'name': 'babu', 'batch': 47, 'course': 'Python', 'age': 50}
print(details.get('name'))
print(details.get('Percentage'))
print(details.get('Percentage',90))
```

```python
# updating or adding items to dictionary
# update(new_dictionary):It updates with new_dictionary items 
details = {'name': 'babu', 'batch': 47, 'course': 'Python', 'age': 50, 'percentage':85}
marks = {"python":90, 'soft skills':85, 'aptitude':78,'percentage':90}
details.update(marks)
print(details)
print(marks)
```

```python
# dictionary not allows duplicate keys
details = {'name': 'babu', 'batch': 47, 'course': 'Python', 'age': 50, 'percentage':85, 'course': 'java'}
print(details)
```

```python
# get dictionary keys and values
# keys() - It returns list of all dictionary keys
# values() - It returns all dictionary items as list of tuples of key-value pairs
details = {'name':'babu', 'percentage':95, 'batch':47, 'course':'python', 'age':50, 'percentage':85}
print(details.keys())
print(details.values())
print(details.items())
print(type(details.keys()))
print(type(details.values()))
print(type(details.items()))
```

```python
details = {'name':'babu', 'percentage':95, 'batch':47, 'course':'python', 'age':50, 'percentage':85}
print(details.keys())
print(details.values())
print(details.items())
print(type(details.keys()))
print(type(details.values()))
print(type(details.items()))
print(list(details.keys()))
print(tuple(details.keys()))
```

```python
# deleting items from dictionary
# pop(key) - It return values and removes the specified key items from dictionary
# popitem() - It returns and removes the last item from dictionary
details = {'name': 'babu', 'batch': 47, 'course': 'Python', 'age': 50}
poped_value = details.pop('age')
print(poped_value)
print(details)
```

```python
details = {'name': 'babu', 'batch': 47, 'course': 'Python', 'age': 50}
poped_value = details.pop('age')
print(poped_value)
print(details)
poped_value = details.pop('age',100)
print(poped_value)
print(details)
item = details.popitem()
print(item)
print(details)
```

```python
# iterate dictionary items
details = {'name':'babu', 'percentage':95, 'batch':47, 'course':'python', 'age':50, 'percentage':85}
for item in details.items():
    print(f"{item[0]}:{item[1]}")
```

```python
# membership checking
details = {'name': 'babu', 'batch': 47, 'course': 'Python', 'age': 50}
key = "python"
if key in details:
    print("key is present")
else:
    print("key is not present")
```

### Creating contact book

```python
###### ph_book={}
def add_contact(name:str,number:int):
    # check wether name is exists in ph_book or not
    if name not in ph_book:
        ph_book[name] = number
        print("Contact saved")
    else:
        print("Name Already exist")

# update mobile number in ph_book
def update_number(name: str, number: int):
    if name not in ph_book:
        print("Contact not exist")
    else:
        ph_book.update({name:number}) # updating mobile number
        print("Mobile number updated")

# delete contact
def delete_contact(name:str):
    if name in ph_book:
        ph_book.pop(name)
        print("Successfully deleted")
    else:
        print("Contact not found")

# Get contact number by name
def get_number(name:str):
        number = ph_book.get(name, 'Name not found')
        print("The contact number updated with:", number) 

# print all contacts
def all_contacts():
    # if contact book is empty
    if not ph_book:
        print("No contacts found")
        return
    for name, number in ph_book.items():
        print(f'{name} : {number}')

# main
if __name__ == '__main__':
    print("welcome to contact book")
    while True:
        print("\n1. Add contact")
        print("2. Update contact")
        print("3. Delete contact")
        print("4. Search contact")
        print("5. Show all contacts")  
        print("6. Exit")

        choice = input("Enter your choice: ")
        if choice == "1":
            name = input("Enter name: ")
            number = int(input("Enter number: "))
            add_contact(name, number)

        elif choice == "2":
            name = input("Enter name: ")
            number = int(input("Enter new number: "))
            update_number(name, number)

        elif choice == "3":
            name = input("Enter name: ")
            delete_contact(name)

        elif choice == "4":
            name = input("Enter name: ")
            get_number(name)

        elif choice == "5":
            all_contacts()
            
        elif choice == "6":
            print("Exit!")
            break
        else:
            print("Invalid choice")
        
```
