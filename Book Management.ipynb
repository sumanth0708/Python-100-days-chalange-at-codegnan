```python
books = {}

def add_book():
    book_id = int(input("Enter Book ID: "))
    name = input("Enter Book name: ")
    author = input("Enter author: ")
    qty = int(input("Enter quantity: "))

    books[book_id] = {"name": name, "author": author, "qty": qty}
    print("Book added!")
def search_book():
    book_id = int(input("Enter ID to search: "))
    if book_id in books:
        print(books[book_id])
    else: 
        print("Book not found")
def delete_book():
    book_id = int(input("Enter ID to delete: "))
    if book_id in books:
        del books[book_id]
        print("Book deleted")
    else:
        print("Book not found")
def total_books():
    print("Total books:", len(books))
def all_books():
    for i, b in books.items():
        print(i, b)

while True:
    print("\n1.Add Book 2.Search Book 3.Delete the book 4.Total books 5.All books 6.Exit")
    choice = input("Choose: ")

    if choice == "1":
        add_book()
    elif choice == "2":
        search_book()
    elif choice == "3":
        delete_book()
    elif choice == "4":
        total_books()
    elif choice == "5":
        all_books()
    elif choice == "6":
        print("HAVE A NICE DAY")
        break
```
