```python
# set allows unique elements
# it does not allows dublicate values
# and it is unordered
s1 = {1,2,3,4,'hi',1,2,3,'a'}
print(s1)
```

```python
# set does not allows list as a element
s1 = {1,2,3,4,'hi',1,2,3,'a', [1,2,3,4]}
print(s1)
```

```python
# set does not allows dicctionary as element
s1 = {1,2,3,4,'hi',1,2,3,'a', {1:2}}
print(s1)
```

```python
## set functions
# Adding elements to set
# add(ele): It add the element to set if element is not in set otherwise nothing
# update(iterator): It add the all elements in iterator
```

```python
s1 = {1,2,3,4}
s1.add(10)
s1.add(1)
print(s1)
s1.update([5,3,5])
print(s1)
```

```python
s1 = {1,2,3,4}
s1.pop()
print(s1)
s1.remove(2)
print(s1)
s1.discard(3)
print(s1)
```

```python
s1 = {1,2,3,5,4}
s2 = {2,3,6,7}
print(s1.union(s2))
print(s1.intersection(s2))
print(s1.difference(s2))
print(s1.symmetric_difference(s2))
print({1}.issubset(s1))
print({6}.issubset(s1))
print({1}.issubset(s2))
```
