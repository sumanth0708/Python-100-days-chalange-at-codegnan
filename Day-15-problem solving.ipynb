## 3.Reverse of number

```python
n = 345
rev = 0
while n>0:
    digit = n % 10
    rev = rev*10+digit
    n = n//10
print(rev)
```

## 4.Armstrong number

```python
num = int(input("Enter a number: "))
digits = len(str(num))
result = 0
for d in str(num):
    result = result + (int(d) ** digits)
if result == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
